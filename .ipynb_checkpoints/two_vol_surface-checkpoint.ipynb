```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import griddata, RBFInterpolator
import seaborn as sns
```

```python
calls_df = pd.read_csv('spy_calls_liquid.csv')
puts_df = pd.read_csv('spy_puts_liquid.csv')

# ATM strikes give (approx) current spot prices 
current_price = calls_df['strike'].median()  
print(f"using spot price: {current_price}")

# clean
calls_clean = calls_df.dropna(subset=['computed_iv']).copy()
puts_clean = puts_df.dropna(subset=['computed_iv']).copy()

print(f"\nCalls: {len(calls_clean)} valid points")
print(f"Puts: {len(puts_clean)} valid points")
```

```python
# compute moneyness, time (yrs)
calls_clean['moneyness'] = calls_clean['strike'] / current_price
calls_clean['T'] = calls_clean['dte'] / 365.0

puts_clean['moneyness'] = puts_clean['strike'] / current_price
puts_clean['T'] = puts_clean['dte'] / 365.0
```

```python
# grid for calls 
n_points = 500 # finer grain
moneyness_min = max(0.85, calls_clean['moneyness'].min())
moneyness_max = min(1.15, calls_clean['moneyness'].max())
T_min = calls_clean['T'].min()
T_max = calls_clean['T'].max()

moneyness_grid = np.linspace(moneyness_min, moneyness_max, n_points)
T_grid = np.linspace(T_min, T_max, n_points)
M, T = np.meshgrid(moneyness_grid, T_grid, indexing = 'ij')

# cubic interpolation 
print("\nInterpolating surface...")
IV_surface_calls = griddata(
    points=(calls_clean['moneyness'], calls_clean['T']),
    values=calls_clean['computed_iv'],
    xi=(M, T),
    method='cubic'
)


# grid for puts
n_points = 500
moneyness_min = max(0.85, puts_clean['moneyness'].min())
moneyness_max = min(1.15, puts_clean['moneyness'].max())
T_min = puts_clean['T'].min()
T_max = puts_clean['T'].max()

moneyness_grid = np.linspace(moneyness_min, moneyness_max, n_points)
T_grid = np.linspace(T_min, T_max, n_points)
M, T = np.meshgrid(moneyness_grid, T_grid)

# cubic interpolation 
print("\nInterpolating put surface...")
IV_surface_puts = griddata(
    points=(puts_clean['moneyness'], puts_clean['T']),
    values=puts_clean['computed_iv'],
    xi=(M, T),
    method='cubic'
)
```

```python
fig = plt.figure(figsize=(20, 10))

# call vol surface
ax1 = fig.add_subplot(2, 3, 1, projection='3d')
surf = ax1.plot_surface(M, T, IV_surface_calls, cmap='viridis', alpha=0.8, edgecolor='none')
ax1.scatter(calls_clean['moneyness'], calls_clean['T'], calls_clean['computed_iv'], c='red', s=5, alpha=0.5)
ax1.set_xlabel('Moneyness (K/S)', fontsize=10)
ax1.set_ylabel('Time to Expiry (years)', fontsize=10)
ax1.set_zlabel('Implied Volatility', fontsize=10)
ax1.set_title('Call Volatility Surface', fontsize=12, fontweight='bold')
ax1.view_init(elev=25, azim=45)
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=5)

# call vol smile (single expiry)
ax2 = fig.add_subplot(2, 3, 2)
# Pick a middle expiry with good data
expiry_call = sorted(calls_clean['expiry'].unique())[2]  # 3rd expiry
exp_data_call = calls_clean[calls_clean['expiry'] == expiry_call].sort_values('moneyness')
dte_call = exp_data_call['dte'].iloc[0]
ax2.plot(exp_data_call['moneyness'], exp_data_call['computed_iv'], 
        marker='o', linewidth=2.5, markersize=6, color='darkblue')
ax2.axvline(x=1.0, color='gray', linestyle='--', linewidth=1, alpha=0.5)
ax2.set_xlabel('Moneyness (K/S)', fontsize=10)
ax2.set_ylabel('Implied Volatility', fontsize=10)
ax2.set_title(f'Call Volatility Smile\n{expiry_call} ({dte_call} DTE)', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3)

# call vol smiles (multiple expiries)
ax3 = fig.add_subplot(2, 3, 3)
expiries_to_plot = sorted(calls_clean['expiry'].unique())[:5]  # first 5
for expiry in expiries_to_plot:
    exp_data = calls_clean[calls_clean['expiry'] == expiry].sort_values('moneyness')
    if len(exp_data) > 3:
        dte = exp_data['dte'].iloc[0]
        ax3.plot(exp_data['moneyness'], exp_data['computed_iv'], 
                marker='o', label=f"{expiry}", linewidth=2, markersize=4)
ax3.axvline(x=1.0, color='gray', linestyle='--', linewidth=1, alpha=0.5)
ax3.set_xlabel('Moneyness (K/S)', fontsize=10)
ax3.set_ylabel('Implied Volatility', fontsize=10)
ax3.set_title('Call Volatility Smiles (Multiple Expiries)', fontsize=12, fontweight='bold')
ax3.legend(fontsize=8)
ax3.grid(True, alpha=0.3)

# put vol surface 
ax4 = fig.add_subplot(2, 3, 4, projection='3d')
surf2 = ax4.plot_surface(M, T, IV_surface_puts, cmap='plasma', 
                         alpha=0.8, edgecolor='none')
ax4.scatter(puts_clean['moneyness'], puts_clean['T'], 
           puts_clean['computed_iv'], c='blue', s=5, alpha=0.5)
ax4.set_xlabel('Moneyness (K/S)', fontsize=10)
ax4.set_ylabel('Time to Expiry (years)', fontsize=10)
ax4.set_zlabel('Implied Volatility', fontsize=10)
ax4.set_title('Put Volatility Surface', fontsize=12, fontweight='bold')
ax4.view_init(elev=25, azim=45)
fig.colorbar(surf2, ax=ax4, shrink=0.5, aspect=5)

# put vol smile (single expiry)
ax5 = fig.add_subplot(2, 3, 5)
# Use same expiry as call smile for comparison
expiry_put = sorted(puts_clean['expiry'].unique())[2]  # 3rd expiry
exp_data_put = puts_clean[puts_clean['expiry'] == expiry_put].sort_values('moneyness')
dte_put = exp_data_put['dte'].iloc[0]
ax5.plot(exp_data_put['moneyness'], exp_data_put['computed_iv'], 
        marker='o', linewidth=2.5, markersize=6, color='darkred')
ax5.axvline(x=1.0, color='gray', linestyle='--', linewidth=1, alpha=0.5)
ax5.set_xlabel('Moneyness (K/S)', fontsize=10)
ax5.set_ylabel('Implied Volatility', fontsize=10)
ax5.set_title(f'Put Volatility Smile\n{expiry_put} ({dte_put} DTE)', fontsize=12, fontweight='bold')
ax5.grid(True, alpha=0.3)

# call vs put IV comparison
ax6 = fig.add_subplot(2, 3, 6)
merged = pd.merge(
    calls_clean[['strike', 'expiry', 'computed_iv', 'moneyness']],
    puts_clean[['strike', 'expiry', 'computed_iv']],
    on=['strike', 'expiry'],
    suffixes=('_call', '_put')
)
ax6.scatter(merged['computed_iv_call'], merged['computed_iv_put'], 
           alpha=0.5, c=merged['moneyness'], cmap='RdYlGn', s=30)
ax6.plot([0, merged[['computed_iv_call', 'computed_iv_put']].max().max()],
         [0, merged[['computed_iv_call', 'computed_iv_put']].max().max()],
         'k--', linewidth=1, alpha=0.5)
ax6.set_xlabel('Call IV', fontsize=10)
ax6.set_ylabel('Put IV', fontsize=10)
ax6.set_title('Call vs Put IV (colored by moneyness)', fontsize=12, fontweight='bold')
ax6.grid(True, alpha=0.3)
cbar = fig.colorbar(ax6.collections[0], ax=ax6)
cbar.set_label('Moneyness', fontsize=8)

plt.tight_layout()
plt.savefig('volatility_surface_complete.png', dpi=300, bbox_inches='tight')
print("\nSaved: volatility_surface_complete.png")
plt.show()
```

```python
fig2, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# atm vol term structure 
atm_calls = calls_clean[
    (calls_clean['moneyness'] > 0.98) & 
    (calls_clean['moneyness'] < 1.02)
].groupby('dte')['computed_iv'].mean().reset_index()

ax1.plot(atm_calls['dte'], atm_calls['computed_iv'], 
         'o-', linewidth=2, markersize=8, color='darkblue')
ax1.set_xlabel('Days to Expiry', fontsize=11)
ax1.set_ylabel('ATM Implied Volatility', fontsize=11)
ax1.set_title('ATM Volatility Term Structure (Calls)', fontsize=13, fontweight='bold')
ax1.grid(True, alpha=0.3)

# skew analysis (25-delta risk reversal proxy)
skew_data = []
for expiry in calls_clean['expiry'].unique():
    exp_calls = calls_clean[calls_clean['expiry'] == expiry]
    exp_puts = puts_clean[puts_clean['expiry'] == expiry]
    
    if len(exp_calls) > 5 and len(exp_puts) > 5:
        # OTM put (moneyness ~ 0.95)
        otm_put = exp_puts[
            (exp_puts['moneyness'] > 0.93) & 
            (exp_puts['moneyness'] < 0.97)
        ]['computed_iv'].mean()
        
        # OTM call (moneyness ~ 1.05)
        otm_call = exp_calls[
            (exp_calls['moneyness'] > 1.03) & 
            (exp_calls['moneyness'] < 1.07)
        ]['computed_iv'].mean()
        
        if not np.isnan(otm_put) and not np.isnan(otm_call):
            skew = otm_put - otm_call  # positive = put skew
            dte = exp_calls['dte'].iloc[0]
            skew_data.append({'dte': dte, 'skew': skew})

skew_df = pd.DataFrame(skew_data).sort_values('dte')
ax2.plot(skew_df['dte'], skew_df['skew'], 
         'o-', linewidth=2, markersize=8, color='darkred')
ax2.axhline(y=0, color='gray', linestyle='--', linewidth=1)
ax2.set_xlabel('Days to Expiry', fontsize=11)
ax2.set_ylabel('Skew (OTM Put IV - OTM Call IV)', fontsize=11)
ax2.set_title('Volatility Skew Term Structure', fontsize=13, fontweight='bold')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('term_structure_analysis.png', dpi=300, bbox_inches='tight')
print("Saved: term_structure_analysis.png")
plt.show()
```

```python
# save everything! 

print("Type of IV_surface_calls:", type(IV_surface_calls))
print("Shape:", IV_surface_calls.shape)

IV_surface_calls = pd.DataFrame(
    IV_surface_calls,
    index=moneyness_grid,
    columns=T_grid
)

IV_surface_puts = pd.DataFrame(
    IV_surface_puts,
    index=moneyness_grid,
    columns=T_grid
)

# Verify no duplicates
print("\nVerifying:")
print("Call surface shape:", IV_surface_calls.shape)
print("Moneyness grid length:", len(moneyness_grid))
print("T grid length:", len(T_grid))
print("Any duplicate moneyness?", pd.Series(moneyness_grid).duplicated().any())
print("Any duplicate times?", pd.Series(T_grid).duplicated().any())

# Save
IV_surface_calls.to_pickle('IV_surface_calls.pkl')
IV_surface_puts.to_pickle('IV_surface_puts.pkl')
```
