```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt
from dataclasses import dataclass
import import_ipynb
from typing import Dict, Tuple
import three_arb_check
black_scholes = three_arb_check.black_scholes
from one_iv_computation import get_spot_and_risk_free_rate 


IV_surface_calls = pd.read_pickle('IV_surface_calls.pkl')
IV_surface_puts = pd.read_pickle('IV_surface_puts.pkl')
```

```python
def calculate_greeks(S, K, T, r, sigma, option_type='call'):
    if T <= 0:
        return {'delta': 0, 'gamma': 0, 'vega': 0, 'theta': 0, 'rho': 0}
    
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    # delta
    if option_type == 'call':
        delta = norm.cdf(d1)
    else:
        delta = norm.cdf(d1) - 1
    
    # gamma
    gamma = norm.pdf(d1) / (S * sigma * np.sqrt(T))
    
    # vega  
    vega = S * norm.pdf(d1) * np.sqrt(T) / 100
    
    # Theta 
    if option_type == 'call':
        theta = (-S * norm.pdf(d1) * sigma / (2 * np.sqrt(T)) 
                 - r * K * np.exp(-r * T) * norm.cdf(d2)) / 365
    else:
        theta = (-S * norm.pdf(d1) * sigma / (2 * np.sqrt(T)) 
                 + r * K * np.exp(-r * T) * norm.cdf(-d2)) / 365
    
    # rho 
    if option_type == 'call':
        rho = K * T * np.exp(-r * T) * norm.cdf(d2) / 100
    else:
        rho = -K * T * np.exp(-r * T) * norm.cdf(-d2) / 100
    
    return {
        'delta': delta,
        'gamma': gamma,
        'vega': vega,
        'theta': theta,
        'rho': rho
    }
```

```python
@dataclass
class Option:
    strike: float
    dte: int  # days to expiration
    option_type: str  # 'call' or 'put'
    
    @property
    def T(self):
        return self.dte / 365.0
    
    @property
    def id(self):
        return f"{self.option_type}_{self.strike:.0f}_{self.dte}"
```

```python
def grid_surface_to_mm_table(iv_grid_df, spot, T_in_years=True, dte_per_year=365,
                            strike_round=1, dte_round=1):
    """
    iv_grid_df: DataFrame indexed by moneyness, columns=T, values=iv
    spot: current underlying spot price
    """
    long = (
        iv_grid_df
        .stack(dropna=True)
        .rename("iv")
        .reset_index()
        .rename(columns={iv_grid_df.index.name or "level_0": "moneyness",
                         iv_grid_df.columns.name or "level_1": "T"})
    )

    # moneyness -> strike
    long["strike"] = long["moneyness"] * spot

    #  T -> dte
    if T_in_years:
        long["dte"] = long["T"] * dte_per_year
    else:
        long["dte"] = long["T"]

    # basic rounding
    long["strike"] = (long["strike"] / strike_round).round() * strike_round
    long["dte"]    = (long["dte"]    / dte_round).round()    * dte_round

    long["strike"] = long["strike"].astype(int)
    long["dte"]    = long["dte"].astype(int)

    return long[["strike", "dte", "iv"]].dropna().drop_duplicates()
```

```python
class MarketMaker:
    """
    quotes bid/ask spreads based on:
    - theo 
    - inventory skew
    - Greek risk penalties
    """
    
    def __init__(self, call_surface_df, put_surface_df, spot_price, risk_free_rate=0.05):
        """
        - call_surface_df: DataFrame with columns [strike, dte, iv] for calls
        - put_surface_df: DataFrame with columns [strike, dte, iv] for puts
        - spot_price: Current underlying price
        - risk_free_rate: Risk-free interest rate
        """
        self.call_surface = call_surface_df
        self.put_surface = put_surface_df
        self.spot_price = spot_price
        self.risk_free_rate = risk_free_rate
        self.inventory = {}  # {option_id: position}
        
        # risk parameters 
        self.base_spread = 0.02  # 2% spread
        self.inventory_skew_factor = 0.005  # 0.5% per contract
        self.delta_limit = 100  # Max delta exposure
        self.vega_limit = 50   # Max vega exposure
        self.delta_penalty_rate = 0.01  # Penalty per unit delta exposure
        self.vega_penalty_rate = 0.02   # Penalty per unit vega exposure
    
    def get_iv(self, option, max_strike_diff=5, max_dte_diff=2):
        if option.option_type == 'call':
            surface = self.call_surface
        else:
            surface = self.put_surface
    
        mask = (surface['strike'] == option.strike) & (surface['dte'] == option.dte)
        matches = surface.loc[mask, 'iv']
        if len(matches) > 0:
            return float(matches.iloc[0])
    
        else:
            print("no matches found for this strike and DTE") 
    
    def get_theo(self, option):
        iv = self.get_iv(option)
        return black_scholes(
            self.spot_price, option.strike, option.T, 
            self.risk_free_rate, iv, option.option_type
        )
    
    def get_greeks(self, option):
        iv = self.get_iv(option)
        return calculate_greeks(
            self.spot_price, option.strike, option.T,
            self.risk_free_rate, iv, option.option_type
        )
    
    def get_portfolio_greeks(self):
        total_delta = 0
        total_gamma = 0
        total_vega = 0
        total_theta = 0
        
        for opt_id, position in self.inventory.items():
            # parse option from id (format: "call_100_30" or "put_95_60")
            parts = opt_id.split('_')
            option = Option(
                strike=float(parts[1]),
                dte=int(parts[2]),
                option_type=parts[0]
            )
            
            greeks = self.get_greeks(option)
            total_delta += position * greeks['delta'] * 100  # per contract
            total_gamma += position * greeks['gamma'] * 100
            total_vega += position * greeks['vega'] * 100
            total_theta += position * greeks['theta'] * 100
        
        return {
            'delta': total_delta,
            'gamma': total_gamma,
            'vega': total_vega,
            'theta': total_theta
        }
    
    def compute_delta_penalty(self, option):
        greeks = self.get_greeks(option)
        portfolio_greeks = self.get_portfolio_greeks()
        
        current_delta = portfolio_greeks['delta']
        option_delta = greeks['delta'] * 100  # per contract
        
        # increases as we approach limits
        if abs(current_delta) > self.delta_limit * 0.7:
            exposure_ratio = abs(current_delta) / self.delta_limit
            penalty = self.delta_penalty_rate * abs(option_delta) * exposure_ratio
            return penalty
        
        return 0
    
    def compute_vega_penalty(self, option):
        greeks = self.get_greeks(option)
        portfolio_greeks = self.get_portfolio_greeks()
        
        current_vega = portfolio_greeks['vega']
        option_vega = greeks['vega'] * 100  # per contract
        
        # increases as we approach limits
        if abs(current_vega) > self.vega_limit * 0.7:
            exposure_ratio = abs(current_vega) / self.vega_limit
            penalty = self.vega_penalty_rate * abs(option_vega) * exposure_ratio
            return penalty
        
        return 0
    
    def quote(self, option):
        """
        generate bid/ask quotes for an option.
        Returns: (bid, ask, theo, components dict)
        """
        theo = self.get_theo(option)
        
        # base spread
        half_spread = self.base_spread * theo
        
        # inventory skew: if long, lower quotes to sell; if short, raise to buy
        position = self.inventory.get(option.id, 0)
        skew = -self.inventory_skew_factor * theo * position
        
        # risk penalties
        delta_penalty = self.compute_delta_penalty(option)
        vega_penalty = self.compute_vega_penalty(option)
        
        # final quotes
        bid = theo - half_spread + skew - delta_penalty
        ask = theo + half_spread + skew + vega_penalty
        
        components = {
            'theo': theo,
            'half_spread': half_spread,
            'inventory_skew': skew,
            'delta_penalty': delta_penalty,
            'vega_penalty': vega_penalty
        }
        
        return bid, ask, theo, components
    
    def trade(self, option, side, quantity=1):
        """
        execute a trade and update inventory.
        
        side: 'buy' or 'sell' from market maker's perspective
        """
        if side == 'buy':
            self.inventory[option.id] = self.inventory.get(option.id, 0) + quantity
        else:
            self.inventory[option.id] = self.inventory.get(option.id, 0) - quantity
    
    def get_inventory_summary(self):
        if not self.inventory:
            return pd.DataFrame()
        
        rows = []
        for opt_id, position in self.inventory.items():
            if position == 0:
                continue
                
            parts = opt_id.split('_')
            option = Option(
                strike=float(parts[1]),
                dte=int(parts[2]),
                option_type=parts[0]
            )
            
            theo = self.get_theo(option)
            greeks = self.get_greeks(option)
            
            rows.append({
                'Option': opt_id,
                'Position': position,
                'Strike': option.strike,
                'DTE': option.dte,
                'Theo': theo,
                'Delta': greeks['delta'] * position,
                'Vega': greeks['vega'] * position,
                'Value': theo * position * 100
            })
        
        df = pd.DataFrame(rows)
        if len(df) > 0:
            df = df.sort_values('Option')
        return df
    
    def visualize_quotes(self, dte, n_strikes=10):
        # get available strikes for this DTE
        call_strikes = self.call_surface[self.call_surface['dte'] == dte]['strike'].values
        put_strikes = self.put_surface[self.put_surface['dte'] == dte]['strike'].values
        strikes = sorted(set(call_strikes) | set(put_strikes))
        
        if len(strikes) == 0:
            print(f"No data for DTE={dte}")
            return
        
        # select strikes around ATM
        atm_idx = np.argmin(np.abs(np.array(strikes) - self.spot_price))
        start = max(0, atm_idx - n_strikes // 2)
        end = min(len(strikes), start + n_strikes)
        strikes = strikes[start:end]
        
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
        
        # calls
        call_bids, call_asks, call_theos = [], [], []
        for strike in strikes:
            try:
                opt = Option(strike, dte, 'call')
                bid, ask, theo, _ = self.quote(opt)
                call_bids.append(bid)
                call_asks.append(ask)
                call_theos.append(theo)
            except:
                call_bids.append(np.nan)
                call_asks.append(np.nan)
                call_theos.append(np.nan)
        
        ax1.plot(strikes, call_theos, 'g--', label='Theo', linewidth=2)
        ax1.plot(strikes, call_bids, 'b-', label='Bid', linewidth=2)
        ax1.plot(strikes, call_asks, 'r-', label='Ask', linewidth=2)
        ax1.fill_between(strikes, call_bids, call_asks, alpha=0.2, color='gray')
        ax1.axvline(x=self.spot_price, color='black', linestyle='--', alpha=0.5, label='Spot')
        ax1.set_xlabel('Strike')
        ax1.set_ylabel('Price ($)')
        ax1.set_title(f'Call Quotes (DTE={dte})')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # puts
        put_bids, put_asks, put_theos = [], [], []
        for strike in strikes:
            try:
                opt = Option(strike, dte, 'put')
                bid, ask, theo, _ = self.quote(opt)
                put_bids.append(bid)
                put_asks.append(ask)
                put_theos.append(theo)
            except:
                put_bids.append(np.nan)
                put_asks.append(np.nan)
                put_theos.append(np.nan)
        
        ax2.plot(strikes, put_theos, 'g--', label='Theo', linewidth=2)
        ax2.plot(strikes, put_bids, 'b-', label='Bid', linewidth=2)
        ax2.plot(strikes, put_asks, 'r-', label='Ask', linewidth=2)
        ax2.fill_between(strikes, put_bids, put_asks, alpha=0.2, color='gray')
        ax2.axvline(x=self.spot_price, color='black', linestyle='--', alpha=0.5, label='Spot')
        ax2.set_xlabel('Strike')
        ax2.set_ylabel('Price ($)')
        ax2.set_title(f'Put Quotes (DTE={dte})')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
```

```python
# USER: INPUT CUSTOM PARAMETERS HERE 
strike = 695
dte = 2
option_type = "call" 
order = "sell" 
quantity = 5
```

```python
spot, rfr = get_spot_and_risk_free_rate()

call_surface_df = grid_surface_to_mm_table(IV_surface_calls, spot)
put_surface_df  = grid_surface_to_mm_table(IV_surface_puts,  spot)


mm = MarketMaker(
       call_surface_df,
       put_surface_df,
       spot,
       rfr
)

# get a quote for your option 
option = Option(strike, dte, option_type)
bid, ask, theo, components = mm.quote(option)
print(f"Bid: ${bid:.2f}, Ask: ${ask:.2f}, Theo: ${theo:.2f}")

mm.trade(option, order, quantity) 
print(mm.get_inventory_summary())
print(mm.get_portfolio_greeks())

mm.visualize_quotes(dte=8, n_strikes=10)
```
